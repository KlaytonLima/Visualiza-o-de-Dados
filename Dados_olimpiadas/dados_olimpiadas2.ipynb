```python
# Importando bibliotecas
import pandas as pd
```

```python
# Importando CSV
df_olimpiadas = pd.read_csv('dataset_olympics.csv')
df_regiao = pd.read_csv('noc_region.csv')
```

```python
# Exibindo as primeiras linhas de um dataframe com head()
# head(). exibe as primeiras linhas de um conjunto de dados
print("Data Frame Olimpiadas")
display(df_olimpiadas.tail())
```

```python
# Exibindo as últimas linhas de um dataframe com tail()
# tail(). exibe as últimas linhas de um conjunto de dados
print("Data Frame Região")
display(df_regiao.tail())
```

```python
# Exibindo quantidade de linhas e colunas de um dataframe com shape
# shape. exibe a quantidae de linhas e colunas de um conjunto de dados
display(df_olimpiadas.shape)
display(df_regiao.shape)
```

```python
# Apresentar estatísticas básicas das colunas do dataframe com describe()
df_regiao.describe()
# top: Valor mais frequente
# freq: Somente dados categóricos
# Dados categórico são dados que pode ser agrupado em categorias em vez de medido numericamente
```

```python
# Apresentar estatísticas básicas das colunas do dataframe com describe()
df_olimpiadas.describe()
# Quartis 25%, 50%, 75% - quatro faixas 
# Dados númericos 
```

```python
# Apresentando todos os campos do dataframe
df_olimpiadas.describe(include='all')
```

```python
df_regiao.describe(include='all')
```

```python
# Exibindo alguns campos do dataframe
df_olimpiadas[['Age', 'Height', 'Weight']].describe()
```

```python
# Listar todos os campos e colunas de um dataframe
list(df_olimpiadas)
```

```python
# Listar os valores didtintos de um campo
df_olimpiadas['City'].unique()
```

```python
# LOC (location): por meio de operadores relacionais(<,>,==,!=) seleciona parte dos dados do dataframe
df_olimpiadas.loc[df_olimpiadas['Team'] == 'Brazil'].head()
```

```python
# LOC: usando operadores lógicos (&, |, ~) para incluir mais de uma condição
# AND(&), OR(|), NOT(~)
df_olimpiadas[(df_olimpiadas['Team'] == 'Brazil') & (df_olimpiadas['Medal'] == 'Gold') & (df_olimpiadas['Year'] == 2016)]
```

```python
# Loc: Usando operadores lógicos (&, |, ~) para incluir mais de uma condição
# AND(&), OR(|), NOT(~)
df_olimpiadas[(df_olimpiadas['Team'] == 'Brazil') & (df_olimpiadas['Medal'] == 'Gold') & (df_olimpiadas['Year'] == 2016) & ~(df_olimpiadas['Sport'] == 'Volleyball')]
```

```python
# Slice: Método que seleciona parte da string de um campo
# Extrair o ano do campo games
df_olimpiadas['Games'].str.slice(start=0, stop=4, step=1)
```

```python
# Slice: Como critério de seleção
df_olimpiadas[(df_olimpiadas['Team'] == 'Brazil') & (df_olimpiadas['Games'].str.slice(start=0, stop=4, step=1) == '2016') & (df_olimpiadas['Medal'] == 'Gold')]
```

```python
# Count: Contagem simples de registros no dataframe
# Quantas medalhas o Brasil ganhou na história das olimpiadas? (até 2016)
df_olimpiadas[(df_olimpiadas['Team'] == 'Brazil') & (df_olimpiadas['Medal'].notna())] ['Medal'].count()
```

```python
# Munique: Contagem de registros distintos
# Quantas modalidades fazem/fizeram parte dos jogos olimpicos?
df_olimpiadas['Sport'].nunique()
```

```python
# Quais?
df_olimpiadas['Sport'].unique()
```

```python
# Explorar Dados: Sumarização
# MEAN e MEDIAN, média e mediana respectivamente relacionados a idedade
# Média
df_olimpiadas[df_olimpiadas['Team'] == 'Brazil'] ['Height'].mean() 
```

```python
# Mediana
df_olimpiadas[df_olimpiadas['Team'] == 'Brazil'] ['Height'].median()
```

```python
# Maxima
df_olimpiadas[df_olimpiadas['Team'] == 'Brazil'] ['Height'].max()
```

```python
# Minima
df_olimpiadas[df_olimpiadas['Team'] == 'Brazil'] ['Height'].min()
```

```python
# Explorar Dados: Agrupamento
# Groupby: Aplica sumarização de valores usando um ou mais campo como índice
# Agrupando medalhas por 'a'
df_olimpiadas.groupby('NOC') ['NOC'].count()
```

```python
# Groupby Multinível: Considera mais e um campo no índice (hierarquia)
# Agrupando medalhas por país, agrupando por tipos de medalhas
df_olimpiadas.groupby(['NOC', 'Medal'])['NOC'].count()
```

```python
# Explorar Dados: Ordenação
# sord_values: ordenação dos resultados de um dataframe ou de uma seleção
df_olimpiadas.sort_values(by = 'Age', ascending = True).head()
```
