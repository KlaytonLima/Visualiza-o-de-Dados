```python
# Listar os valores didtintos de um campo
df_olimpiadas['City'].unique()
```

```python
# LOC (location): por meio de operadores relacionais(<,>,==,!=) seleciona parte dos dados do dataframe
df_olimpiadas.loc[df_olimpiadas['Team'] == 'Brazil'].head()
```

```python
# LOC: usando operadores lógicos (&, |, ~) para incluir mais de uma condição
# AND(&), OR(|), NOT(~)
df_olimpiadas[(df_olimpiadas['Team'] == 'Brazil') & (df_olimpiadas['Medal'] == 'Gold') & (df_olimpiadas['Year'] == 2016)]
```

```python
# Loc: Usando operadores lógicos (&, |, ~) para incluir mais de uma condição
# AND(&), OR(|), NOT(~)
df_olimpiadas[(df_olimpiadas['Team'] == 'Brazil') & (df_olimpiadas['Medal'] == 'Gold') & (df_olimpiadas['Year'] == 2016) & ~(df_olimpiadas['Sport'] == 'Volleyball')]
```

```python
# Slice: Método que seleciona parte da string de um campo
# Extrair o ano do campo games
df_olimpiadas['Games'].str.slice(start=0, stop=4, step=1)
```

```python
# Slice: Como critério de seleção
df_olimpiadas[(df_olimpiadas['Team'] == 'Brazil') & (df_olimpiadas['Games'].str.slice(start=0, stop=4, step=1) == '2016') & (df_olimpiadas['Medal'] == 'Gold')]
```

```python
# Count: Contagem simples de registros no dataframe
# Quantas medalhas o Brasil ganhou na história das olimpiadas? (até 2016)
df_olimpiadas[(df_olimpiadas['Team'] == 'Brazil') & (df_olimpiadas['Medal'].notna())] ['Medal'].count()
```

```python
# Munique: Contagem de registros distintos
# Quantas modalidades fazem/fizeram parte dos jogos olimpicos?
df_olimpiadas['Sport'].nunique()
```

```python
# Explorar Dados: Sumarização
# MEAN e MEDIAN, média e mediana respectivamente relacionados a idedade
# Média
df_olimpiadas[df_olimpiadas['Team'] == 'Brazil'] ['Height'].mean() 
```

```python
# Mediana
df_olimpiadas[df_olimpiadas['Team'] == 'Brazil'] ['Height'].median()
```

```python
# Maxima
df_olimpiadas[df_olimpiadas['Team'] == 'Brazil'] ['Height'].max()
```

```python
# Minima
df_olimpiadas[df_olimpiadas['Team'] == 'Brazil'] ['Height'].min()
```

```python
# Explorar Dados: Agrupamento
# Groupby: Aplica sumarização de valores usando um ou mais campo como índice
# Agrupando medalhas por 'a'
df_olimpiadas.groupby('NOC') ['NOC'].count()
```

```python
# Groupby Multinível: Considera mais e um campo no índice (hierarquia)
# Agrupando medalhas por país, agrupando por tipos de medalhas
df_olimpiadas.groupby(['NOC', 'Medal'])['NOC'].count()
```

```python
# Explorar Dados: Ordenação
# sort_values: ordenação dos resultados de um dataframe ou de uma seleção
df_olimpiadas.sort_values(by = 'Age', ascending = True).head()
```

```python
# sort_values: Ordenação de dados Agrupados
df_olimpiadas.groupby(['NOC', 'Medal']) ['NOC'].count().sort_values(ascending = False)
```

```python
#sort_values: Persistndo a ordenação do dataframe
df_olimpiadas.sort_values(by = 'Age', ascending = True, inplace = True)
df_olimpiadas.head()
```

```python
# Ex0plorar dados: Indexação
# set_index:
# Quando não especificado na criação do dataframe, o pandas cria um índice default sequencial
# No momento da importação é possível definir um índice Sobre um dataframe já criado
# set_index() define um novo índice
# Não precisa ser necessariamente campo único
df_olimpiadas.set_index(['ID'], inplace = True)
df_olimpiadas.head()
```

```python
# reset_index:
# Inclue o campo de índice como um campo normal e cria um novo índice
# Se não quiser incluir o campo de índice como um campo normal, informar drop=true
df_olimpiadas.reset_index(inplace = True)
df_olimpiadas.head()
```

```python
# Explorar dados: Amostragem
# Amostragem é um tema extenso e complexo
# Trata-se de uma disciplina de cursos de estatística onde técnicas são estudadas
# Aqui será vista a amostragem randomica, ùtil para resumir grand quantidade de dados
# sample: Cria uma amostra aleatória a partir de uma fração (parâmetro frac do método)
df_olimpiadas.sample(frac=0.1).shape
```

```python
# sample:
# Cada vez que for chamado a amostra será diferente
# Para a reptição de um experimento podemos usar uma "semente", por meio do parâmetro random_state
df_olimpiadas.sample(frac=0.1).head()
```

```python
# Repetindo operação termos outros dados
df_olimpiadas.sample(frac=0.1).head()
```

```python
# Fixando uma semente
df_olimpiadas.sample(frac=0.1, random_state = 123).head()
```

```python
# Repetindo teremos os mesmos dados
df_olimpiadas.sample(frac=0.1, random_state = 123).head()
```

```python
# Criando novo dataframe a partir de um sample
df_sample10 = df_olimpiadas.sample(frac=0.1, random_state = 123)
df_sample10.head()
```

```python
# Explorar dados: Gráficos
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
# Instrui o Jupyter Notebook a exibir os gráficos diretamente nas células abaixo.
```

```python
x = np.linspace(0,5,15)
# 0: Valor inicial da sequência
# 5: Valor final da sequência
# 15: Número total de pontos gerados
y = x**2;
display(x)
display(y)
```

```python
# Matplotlib
# Gráfico simples
plt.plot(x,y,'r') 
# 'r' cor vermelha
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
```

```python
# Matplotlib 
# Gráfico simples - eixos X e Y sobrepostos
plt.plot(x, x**2, label = "x^2")
plt.plot(x, x**3, label = "x^3")
```

```python
# Gráfico de distribuição
# Hsiplot - gráfico de distribuição - histograma
sns.histplot(df_olimpiadas['Age'], kde=True)
# kde: Curva de densidade kernel (KDE)
```
